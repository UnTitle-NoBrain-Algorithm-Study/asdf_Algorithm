# 통계학

```python
import pandas as pd
```

```python
_input = int(input())
num_list = []
for i in range(_input):
    num_list.append(int(input()))

df = pd.Series(num_list)
df.describe()
```

```python
mean = round(df.mean())
median = round(df.median())
_range = round(df.max()) - round(df.min())
```

```python
mode_list = list(df.mode())
mode_list
if len(mode_list) > 1:
    mode = mode_list[1]
else:
    mode = mode_list[0]
```

```python
print(mean)
print(median)
print(mode)
print(_range)
```

# 생태학

```python
"""
import sys
input = sys.stdin.readline
"""""
#백준용, sys.stdin.readline 
```

```python
dic = {}
keys = []
num = 0
while 1:
    _input = input().rstrip()
    if bool(_input) == False: # if not _input:
        break
    num +=1
    if _input in dic:
        dic[_input] +=1
    else:
        dic[_input] = 1
        keys.append(_input)
```

```python
dic
```

```python
keys.sort()
keys
```

```python
for i in range(len(keys)):
    print('%s %.4f'%(keys[i],dic[keys[i]]/num*100))
```

# 알약

```python
wh = []
for k in range(31):
    wh.append([0 for i in range(k+1)])
for w in range(31):
    for h in range(len(wh[w])-1):
        if h == 0:
            wh[w][h] = 1
        elif w>1 and h>0:
            wh[w][h] = wh[w][h-1] + wh[w-1][h]

wh
```

```python
sum(wh[30])
```

```python
while 1:
    _input = int(input())
    if _input == 0:
        break;
    print(sum(wh[_input]))
```
